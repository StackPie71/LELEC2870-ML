```python
diab = scipy.io.loadmat('data2/diabetes.mat')
t = diab['t']
diabX = diab['X']

print(np.corrcoef(diabX[:,0],diabX[:,1]))
plt.figure(figsize=(12,12))
plt.subplot2grid((4,3),(0,0))
plt.title('age')
plt.plot(diabX[:,0],t,'o')

plt.subplot2grid((4,3),(0,1))
plt.title('sex')
plt.plot(diabX[:,1],t,'o')

plt.subplot2grid((4,3),(0,2))
plt.title('bmi')
plt.plot(diabX[:,2],t,'o')

plt.subplot2grid((4,3),(1,0))
plt.title('blood_pressure')
plt.plot(diabX[:,3],t,'o')

plt.subplot2grid((4,3),(1,1))
plt.title('serum_1')
plt.plot(diabX[:,4],t,'o')

plt.subplot2grid((4,3),(1,2))
plt.title('serum_2')
plt.plot(diabX[:,5],t,'o')

plt.subplot2grid((4,3),(2,0))
plt.title('serum_3')
plt.plot(diabX[:,6],t,'o')

plt.subplot2grid((4,3),(2,1))
plt.title('serum_4')
plt.plot(diabX[:,7],t,'o')
                 
plt.subplot2grid((4,3),(2,2))
plt.title('serum_5')
plt.plot(diabX[:,8],t,'o')

plt.subplot2grid((4,3),(3,0))
plt.title('serum_6')
plt.plot(diabX[:,9],t,'o')

plt.show()
```
